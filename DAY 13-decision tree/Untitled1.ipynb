```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import accuracy_score, classification_report

# ==========================================
# 1. DATASET LOAD KARNA
# ==========================================
# File ko load karein (ensure karein ki file aapki script ke sath same folder me ho)
file_path = "Decision_Tree_Large_Practice_Dataset.xlsx"
df = pd.read_excel(file_path)

print("--- Dataset Ke Pehle 5 Rows ---")
print(df.head())

# ==========================================
# 2. DATA PRE-PROCESSING (Encoding)
# ==========================================
# Customer_ID humare prediction ke liye zaroori nahi hai, isliye ise drop karenge
X = df.drop(columns=['Customer_ID', 'Made_Purchase'])

# Target Variable ('Yes'/'No') ko binary (1/0) me convert karenge
y = df['Made_Purchase'].map({'Yes': 1, 'No': 0})

# Categorical columns (like Age_Group, Income) ko numerical format me badlenge
X_encoded = pd.get_dummies(X, drop_first=True)

# ==========================================
# 3. TRAIN-TEST SPLIT
# ==========================================
# 80% data training ke liye aur 20% data testing ke liye split karenge
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42)

# ==========================================
# 4. MODEL TRAINING & TUNING
# ==========================================
# Ek Decision Tree Classifier banayein (overfitting rokne ke liye max_depth lagaya hai)
model = DecisionTreeClassifier(criterion='entropy', max_depth=4, random_state=42)
model.fit(X_train, y_train)

# ==========================================
# 5. MODEL EVALUATION (Testing)
# ==========================================
y_pred = model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print("\n--- Model Evaluation ---")
print(f"Accuracy Score: {accuracy * 100:.2f}%")
print("\nClassification Report:")
print(classification_report(y_test, y_pred, target_names=['No', 'Yes']))

# ==========================================
# 6. DECISION TREE VISUALIZATION
# ==========================================
plt.figure(figsize=(16, 8))
plot_tree(
    model, 
    feature_names=X_encoded.columns, 
    class_names=['No', 'Yes'], 
    filled=True, 
    rounded=True, 
    fontsize=10
)
plt.title("Trained Decision Tree Structure", fontsize=14)
plt.show()
```
