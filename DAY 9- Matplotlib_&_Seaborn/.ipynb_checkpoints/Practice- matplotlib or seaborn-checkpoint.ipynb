```python
import matplotlib.pyplot as plt
from matplotlib.patches import Circle

x1=[4,6]
x2=[4,4]
x3=[6,6]
x4=[6,10]
x5=[10,10]
x6=[6,10]
x7=[4,5]
x8=[5,6]

y1=[2,2]
y2=[2,8]
y3=[2,8]
y4=[2,2]
y5=[2,8]
y6=[8,8]
y7=[8,12]
y8=[12,8]

x_door = [4.5, 4.5, 5.5, 5.5]
y_door = [2, 5, 5, 2]

x_window = [7.5, 8.5, 8.5, 7.5, 7.5]
y_window = [4, 4, 6, 6, 4]

plt.plot(x1,y1, color='black')
plt.plot(x2,y2, color='black')
plt.plot(x3,y3, color='black')
plt.plot(x4,y4, color='black')
plt.plot(x5,y5, color='black')
plt.plot(x6,y6, color='black')
plt.plot(x7,y7, color='black')

plt.plot(x8,y8, color='black')
plt.plot(x_door, y_door, color='black')
plt.plot(x_window, y_window, color='black')

plt.plot([7.5, 8.5], [5, 5], color='black')
plt.plot([8, 8], [4, 6], color='black')     
plt.plot([4, 6], [8, 8], color='black')     
plt.plot([5, 9], [12, 12], color='black')   
plt.plot([9, 10], [12, 8], color='black')   
```

```python
import matplotlib.pyplot as plt
from matplotlib.patches import Circle, Polygon, Wedge

fig, ax = plt.subplots(figsize=(6, 6))
ax.axis('on') 

# Gold color ka circle
face_c = Circle([6, 6], 5.3, color='#F2AE16', fill=True, lw=1, ec='#A67F10')
ax.add_patch(face_c)

# Sunglasses
left_lens_points = [[0.7, 8], [5.5, 8], [5.1, 5.5], [1.3, 5.5]]
right_lens_points = [[6.5, 8], [11.3, 8], [10.7, 5.5], [6.9, 5.5]]

left_trapezoid = Polygon(left_lens_points, color='black', fill=True)
right_trapezoid = Polygon(right_lens_points, color='black', fill=True)
ax.add_patch(left_trapezoid)
ax.add_patch(right_trapezoid)

# Sunglasses ka bridge
ax.plot([5.5, 6.5], [8, 8], color='black', lw=6) 

# Chashme par glints
glint_l = Circle([1.8, 7.0], 0.4, color='#808080')
glint_r = Circle([10.2, 7.0], 0.4, color='#808080')
ax.add_patch(glint_l)
ax.add_patch(glint_r)

# Smile
smile_w = Wedge([6, 4.2], 2, theta1=200, theta2=340, color='#8B4513', fill=False, lw=5)
ax.add_patch(smile_w)

plt.show()
```

```python
import seaborn as sns
iris_data= sns.load_dataset('iris')
iris_data.head(6)
```

```python
sns.lineplot(x = 'sepal_length' , y = 'sepal_width' , data = iris_data)
plt.title('line Plot Graph')
plt.show()
```

```python
plt.figure(figsize=(15, 6))
sns.barplot(x = 'sepal_length' , y = 'sepal_width' , data = iris_data,color="green")
```

```python
plt.figure(figsize=(15, 6))
sns.countplot(x = 'sepal_length' , data = iris_data,color="green")
```

```python
plt.figure(figsize=(15, 6))
sns.boxplot(x = 'sepal_length' , y = 'sepal_width' , data = iris_data,color="green")
```

```python
plt.figure(figsize=(15, 6))
sns.violinplot(x = 'sepal_length' , y = 'sepal_width' , data = iris_data,color="green")
```

```python
sns.swarmplot(x = 'species' , y = 'sepal_width' , data = iris_data,color="green")
```

```python
sns.pairplot(iris_data , hue = 'species')
```
