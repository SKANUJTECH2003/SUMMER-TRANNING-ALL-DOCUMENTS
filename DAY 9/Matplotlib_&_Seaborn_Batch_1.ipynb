```python
import matplotlib.pyplot as plt
```

# 1. Line Chart:
## It is used for show the trends over the time.

```python
x = [3,7,5,9]
y = [1,2,3,4]
plt.plot(x,y, marker = 'o')
plt.title("Line Chart")
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid()
plt.show()
```

### Bar Chart/graph :
### It is used for comapring quatities by categories.

```python
x = ['A' , 'B', 'C' , 'D']
y = [35,55,40,60]

plt.bar(x,y , color = 'skyblue')
plt.show()
```

```python
plt.barh(x,y,color = 'green')
plt.show()
```

### Histogram
### It is used for showing the frequency ditribution of a dataset.

```python
import numpy as np

data = np.random.randn(1000)

plt.hist(data , bins = 30 , color = 'red')
plt.grid()
plt.show()
```

### Pie Chart:
### It is used to show the proportions or percentage of data.

```python
subs = ['C++' , 'Java' , 'HTML' , 'Pyhton']
x = [25,40,15,20]

plt.pie(x , labels = subs , autopct = '%1.1f%%' , startangle = 90)
plt.show()
```

### Scatter Plot:
### It used to show the relationship between two variables.

```python
x = [12,4,6,17,13,27,25,41]
y = [89,102,111,95,112,118,109,121]

plt.scatter(x,y)
plt.show()
```

### Box plot (Whisker plot):
### It shows the Statistics summary like median , quartile and outliers.

```python
x = [34,5,45,32,25,30,20,40,50 , 230]     

plt.boxplot(x)
plt.show()
```

```python
### multiple line plot

x = [2,5,6,8,10]
y1 = [7,3,8,10,4]
y2 = [5,10,1,6,8]
y3 = [8,2,5,7,4]

plt.plot(x , y1 , marker ='*' , color = 'green')
plt.plot(x , y2 , marker ='o' , color = 'red')
plt.plot(x , y3 , marker ='+' , color = 'skyblue')

plt.show()
```

# Seaborn :
## It is powerful data visualization python library based on matplotlib. It provides high level and attractive statisticals graphs and plots.

```python
!pip install seaborn 
```

```python
import seaborn as sns
```

```python
tip = sns.load_dataset('tips')
```

```python
tip.head(6)
```

```python
iris = sns.load_dataset('iris')
```

```python
iris.head(6)
```

```python
sns.lineplot(x = 'tip' , y = 'total_bill' , data = tip)
plt.title('line-plot')
plt.show()
```

```python
sns.barplot(x = 'day' , y = 'total_bill' , data = tip)
```

```python
# Countplot : It is used to count the occurences of categorical data.

sns.countplot(x='day' , data = tip)
```

```python
# boxplot
```

```python
sns.boxplot(x = 'day' , y = 'total_bill' , data = tip)
```

```python
# violin plot:
#It is used for distribution and probability density.

sns.violinplot(x = 'day' , y = 'total_bill' , data = tip)
```

```python
# swarm plot: 
#  It is used for display all the data points on categorical axis.

sns.swarmplot(x= 'day' , y = 'total_bill' , data = tip)
```

```python
# pairplot:
#It is used to ploting pairwise relationships of dataset. 

sns.pairplot(iris , hue = 'species')
```

```python
# heatmap :
# It is used to show the corelation of dataset.
coor = tip.corr()
cor = tip.corr(numeric_only = True)
sns.heatmap(cor , annot = True , cmap = 'coolwarm' )
```
